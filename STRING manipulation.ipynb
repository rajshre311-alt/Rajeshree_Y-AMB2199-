```python
a="Today is 20-Jan-2026 MONDAY"
B=V=C=D=O=0
for i in a:
  if i==" ":
    B=B+1
  elif i in "aeiouAEIOU":
    V=V+1
  elif i.isalpha():
    C=C+1
  elif i.isdigit():
    D=D+1
  else:
    O=O+1
print("spaces:",B)
print("vowels:",V)
print("consonents:",C)
print("digits:",D)
print("others:",O)
```
